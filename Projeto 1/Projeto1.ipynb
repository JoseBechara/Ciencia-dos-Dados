```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
energia = pd.read_excel('electricity_use_per_person.xlsx')
internet = pd.read_excel('internet_users.xlsx')
IDH = pd.read_excel('hdi_human_development_index.xlsx')
```

```python
energia_2014 = energia.loc[:,["geo", 2014]]
internet_2014 = internet.loc[:,["geo", 2014]]
IDH_2014 = IDH.loc[:,["geo", 2014]]
```

```python
energia_2014 = energia_2014.set_index("geo")
internet_2014 = internet_2014.set_index("geo")
IDH_2014.rename(columns={"geo":"Country"}, inplace=True)
IDH_2014 = IDH_2014.set_index("Country")
```

```python
df = IDH_2014.join(energia_2014, how='inner', lsuffix = ' IDH', rsuffix = ' Energia por pessoa')
dfjoin = df.join(internet_2014, how = 'inner')
dfjoin.rename(columns={2014:"2014 Usuarios de Internet (%)"}, inplace=True)

dfjoin
```

```python
dfjoin.plot.scatter(x="2014 Energia por pessoa", y = "2014 IDH");

dfjoin.plot.scatter(x="2014 Usuarios de Internet (%)", y = "2014 IDH");

dfjoin.plot.scatter(x="2014 Usuarios de Internet (%)", y = "2014 Energia por pessoa");

```

# Análise dos dados e Conclusão:

### IDH x Energia por pessoa:
#### Sendo o IDH o índice que mede o desenvolvimento humano, é esperado que quanto maior for seu valor, mais desenvolvido é o país e portanto mais pessoas tem acesso a coisas básicas como energia elétrica, é provável que dessa forma, o gráfico mostre o consumo de energia elétrica proporcional ao IDH. No entanto, esse comportamento não é mostrado de forma tão acentuada quanto se espera, isso se deve, possivelmente, pela mudança na forma com que a eletricidade é usada. A eficiência dos equipamentos Eletro-eletronicos é cada vez mais alta, o que diminui o consumo de energia elétrica, como nos países com IDH mais alto, em geral, suas populações possuem um poder aquisitivo maior, assim tendo estes equipamentos mais eficientes.


### IDH x Usuários de internet:
#### O mesmo raciocínio se aplica neste caso, quanto maior for o IDH, mais desolvido é o país, e portanto, mais pessoas tem acesso a serviços como internet. Entretanto, desta vez o gráfico se comporta como o esperado, proporcional ao IDH. A ligação tão direta entre o IDH e a taxa de usuários de internet pode ser explicada pelo fato de que, a internet é uma ferramenta poderosa que basicamente move um país, ela facilita ações em diversas áreas e portanto, quando mais desenvolvido for um país, maior será seu uso.


### Energia por pessoa x Usuários de internet:
#### Apesar da taxa de correlação entre elas nao seja tão expressiva quanto IDH com internet, ela continua sendo significativa. Isso se da por um motivo óbivio: uma depende da outra para existir, sem energia elétrica, não seria possível fazer uso da internet. Isso mostra que, países em desenvolvimento podem ter um alto consumo de energia elétrica e não utilizar a internet da mesma forma.
